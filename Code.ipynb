```python
import pandas as pd
from pandas_datareader import data as pdr
yf.pdr_override()
from datetime import datetime
```

```python
companies = ['MSFT', 'AMZN']
```

```python
end = datetime.now()
start = datetime(end.year, 1, 1)
```

```python
for stocks in companies:
  globals()[stocks] = yf.download(stocks, start, end)
```

```python
company_list = [MSFT, AMZN]
company_name = ["Microsoft", "Amazon"]
```

```python
for company, name in zip(company_list, company_name):
  company["company_name"] = name

df = pd.concat(company_list, axis = 0)
```

```python
from ast import increment_lineno
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
```

```python
plt.figure(figsize=(15, 10))

for i, company in enumerate(company_list, 1):
  plt.subplot(2, 2, i)
  company["Adj Close"].plot()
  plt.ylabel('Adj Close')
  plt.title(f"Closing price of {company_name[i-1]}")

plt.tight_layout()
```

```python
plt.figure(figsize=(15, 10))

for i, company in enumerate(company_list, 1):
  plt.subplot(2, 2, i)
  company["Volume"].plot()
  plt.ylabel('Volume')
  plt.title(f"Sales Volume of {company_name[i-1]}")

plt.tight_layout()
```

```python
ma_day = [7, 14, 30]
for ma in ma_day:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Adj Close'].rolling(ma).mean()

fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_figheight(7)
fig.set_figwidth(14)

MSFT[['Adj Close', 'MA for 7 days', 'MA for 14 days', 'MA for 30 days']].plot(ax=axes[0])
axes[0].set_title('Microsoft')

AMZN[['Adj Close', 'MA for 7 days', 'MA for 14 days', 'MA for 30 days']].plot(ax=axes[1])
axes[1].set_title('Amazon')

fig.tight_layout()
```
